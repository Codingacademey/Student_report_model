```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Lasso,Ridge
from xgboost import XGBRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('exam_score', axis=1), df['exam_score'], test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import VotingRegressor

voting_reg = VotingRegressor(
    estimators=[('lr', LinearRegression()),('ls',Lasso()), ('rf', RandomForestRegressor()),('xgb',XGBRegressor())]
    ,weights=[3, 1, 2, 2]
)
```

```python
from sklearn.ensemble import StackingRegressor

stack_reg = StackingRegressor(
    estimators=[
        ('lr', LinearRegression()),
        ('ls', Lasso()),
        ('rf', RandomForestRegressor()),
        ('xgb', XGBRegressor())
    ],
    final_estimator=LinearRegression()
)
stack_reg.fit(X_train, y_train)
print("Stacking R2:", stack_reg.score(X_test, y_test))
```

```python
pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('reg',stack_reg)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
Y_predict=pipe.predict(X_test)
```

```python
r2_score(y_test,Y_predict)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(stack_reg, X_train, y_train, cv=5, scoring="r2")
print("CV mean R2:", scores.mean())
```
